## **3.2.1 Limits and Continuity**

Limits are essential to calculus and mathematical analysis, and are used to define continuity, derivatives, and integrals. In this chapter, we will use the Euclidean Norm $\|x\|=\sqrt{\Sigma_{i=1}^d x^2_i}$ for $x=(x_i,...,x_d)^T \epsilon$ $R^d.$ The open *r*-ball around **x** $\epsilon$ $R^d$ is the set of points within Euclidean distance $r$ of **x**, that is,
$$B_r(x)=\{y\hspace{1 mm}\epsilon \hspace{1mm} R^d : \|y-x\| < r \}.$$
A point **x** $\epsilon$ $R^d$ is a limit point (or accumulation point) of a set $A \subseteq R^d$ if every open ball around **x** contains an element of **a** of $A$ such that $a \neq x$ as in Figure 3.1. A set $A$ is closed if every limit point of $A$ belongs to $A.$ A set $A$ is open if there is a
$B_r(x) \subseteq A$ for all $x$ $\epsilon$ $A$ as in figure 3.2. A set $A \subseteq R^d$ is bounded if there exists an $r > 0$ such that $A \subseteq B_r(0),$ where $0=(0,...,0)^T.$

Definition 3.2.1 (Limits of a function) Let $f:D\rightarrow R$ be a real-valued function. Then $f$ is said to have a limit $L$ $\epsilon$ $R$ as **x** approaches **a** if: for any $\varepsilon > 0$ such that $|f(x)-L| < \varepsilon$ for all **x** $\epsilon$ $D\cap B_\delta(a)$ $\backslash\{a\}.$
This is written as
$$\lim_{x\to\infty}f(x)=L.$$
Continuous function is a function that does not have any abrupt changes in value, known as as discontinuities. The definition of continuity of a function can be visaulized in Figure 3.3 Note that we explicitly exclude **a** to be $D.$

Definition 3.2.2(Continous function) Let $f:D\to R$ be a real-valued function on $D\subseteq R^d.$ Then $f$ is said to be continuous at $a$ $\epsilon$ $D$ if
$$\lim_{x\to a}f(x)=f(a).$$
Figure 3.3 is an example of continous function. Functions are often obtained from the composition of simpler ones. We will use the standard notation $h=g \circ f$ is continous at $x_0.$
Extremum lemna is very useful. We will not prove it here. It's proof can be found in many books.

Definition 3.2.4 (Extremmum) Suppose $f:D\to R$ is defined on a set $D \subseteq R^d.$ We say that $f$ attains a maximum value $M$ at **z**$^*$ if $f$(**z**$^*$) $\hspace{1mm} =M$ and $M \ge$ $f$(**x**) for all x $\epsilon \hspace{1 mm}$$D.$

Theorem 3.2.5 (Extreme Value) Let $f:D\to R$ be a real-valued, continous function on a nonempty, closed, bounded set $D \subseteq R^d.$ Then $f$ attains a maximum and a minimum of $D.$



## **3.2.2 Derivatives**
### 3.2.2.1 Single-Variable Case

The derivative of a function of a real variable measures the sensitivity to change of the funciton value (output value) with respect to another variable. We begin by reviewing the sign-variable case.

Definition 3.2.6 (Derivative) Let $f:D\to R$ where $D\subseteq R$ and let $x_0$ $\epsilon$ $D$ be an interior point of $D.$ The derivative of $f$ at $x_0$ is
$$f'(x_0)=\frac{df(x_0)}{2}=\lim_{h\to 0}\frac{f(x_0+h)-f(x_0)}{h},$$ provided that the limit exists.

The definition of derivative of a function can be illustrated at in Figure 3.The following proposition indicates that the differentiation is a linear operator.

Proposition 3.2.7 Let $f$ and $g$ have derivatives at $x$ and let $\alpha$ and $\beta$ be constants. The following results hold

$$[\alpha f(x)+\beta g(x)]'=\alpha f'(x)+\beta g'(x).$$

The following lemna encapsulates a key insight about the derivative of $f$ at $x_0:$ it tells us where to find the extreme values.

Lemma 3.2.8 Let $f:D \to R$ with $D \subseteq R$ and let $x_0$ $\epsilon$ $D$ be an interior where $f'(x_0)$ exists. If $f'(x_0) > 0,$ then there is an open ball $B_\delta (x_0):$

$$a. f(x)>f(x_0)\hspace{2mm} if \hspace{2mm} x > x_0,$$
$$b. f(x)<f(x_0)\hspace{2mm} if \hspace{2mm} x < x_0.$$

If instead $f'(x_0) <0,$ the opposite holds.










```python
### Limits and Continuity

## Limits

import sympy as sp

# Example 1: Find the limit of sin(5*x)/x as x goes to zero

x = sp.Symbol('x')
f = sp.sin(5*x)/x
L = sp.limit(f,x,0)

print("Our limit for our example 1 is: ", L)

# Example 2: Find the limit of sin(3*y)/2*y as x goes to zero

y = sp.Symbol('y')
f2 = sp.sin(3*y)/(2*y)
L2 = sp.limit(f2,y,0)

print("Our limit for our example 2 is: ", L2)

## Continuity

import matplotlib.pyplot as plt
import numpy as np

# Example 1: Plot the continous function y = x**2

x = np.arange(1, 11)
y = x**2
plt.title("Graph for y = x^2")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.plot(x,y, 'r')
plt.show()

# Examaple 2: Plot the continus function cos(x)

x = np.arange(0, 2 * np.pi, 0.1)
y = np.cos(x)
plt.title("Graph for y = cos(x)")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.plot(x,y, 'r')
plt.show()

```
